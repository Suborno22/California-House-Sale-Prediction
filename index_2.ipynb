```python
from sklearn.cluster import KMeans
import numpy as np

# 10 houses with (lat, lon) coordinates
houses = np.array([
    [40.73, -74.01],  # Downtown
    [40.72, -74.00],  # Downtown
    [40.71, -74.02],  # Downtown
    [40.76, -73.98],  # Uptown
    [40.75, -73.99],  # Uptown
    [40.77, -73.97],  # Uptown
    [40.68, -73.94],  # Brooklyn
    [40.67, -73.95],  # Brooklyn
    [40.69, -73.93],  # Brooklyn
    [40.66, -73.96],  # Brooklyn
])

# Create 3 clusters
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(houses)

# See which cluster each house belongs to
print("Cluster assignments:", kmeans.labels_)
# Output might be: [0, 0, 0, 1, 1, 1, 2, 2, 2, 2]

# See the 3 cluster centers
print("Cluster centers:")
print(kmeans.cluster_centers_)
```
